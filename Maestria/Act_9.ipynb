```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from IPython.display import Markdown
import warnings
```

```python
data = 'Act9.xlsx'
df = pd.read_excel(data)
df.head()
```

```python
produccion = sns.lmplot(
    data=df,
    x='Piezas',
    y='Costo total de produccion',
    height=7,
    scatter_kws={'s': 100}
)

plt.show()

```

```python
correlacion=df.corr()
correlation_value = round(correlacion.loc['Costo total de produccion', 'Piezas'],2)
print(f"Coeficiente de correlación: {correlation_value}")
```

```python
##Se crea el modelo de regresion lineal
modelo = LinearRegression()
##Se preparan los datos de entrada y salida
x = df[["Piezas"]]
y = df[["Costo total de produccion"]]
##Se ajusta el modelo
modelo.fit(x,y)
##Valor a predecir,costo de 127 cupcakes
x_valor=127
##Obtener los valores
y_predicho=modelo.predict([[x_valor]])
pendiente = round(modelo.coef_[0][0],2)
ordenada_al_origen = round(modelo.intercept_[0],2)

print(f"El costo total de produccion por {x_valor} cupcakes es de: {round(y_predicho[0][0], 2)}")
print(f"La pendiente del modelo es: {pendiente}")
print(f"La ordenada al origen es: {ordenada_al_origen}")
```
