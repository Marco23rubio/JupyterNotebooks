```python
from scipy.stats import f_oneway
```

```python
tratamiento_1 = [9.8, 10.1, 9.8, 11.3, 10.7, 10.7]
tratamiento_2 = [9.2, 8.6, 9.2, 10.3, 10.7, 10.2]
tratamiento_3 = [8.4, 7.9, 8.0, 9.8, 10.1, 10.1]
```

```python
f_statistic, p_value = f_oneway(tratamiento_1, tratamiento_2, tratamiento_3)
```

```python
print("Estadístico F:", f_statistic)
print("Valor p:", p_value)
```

```python
import numpy as np
```

```python
media_tratamiento_1 = round(np.mean(tratamiento_1), 2)
media_tratamiento_2 = round(np.mean(tratamiento_2), 2)
media_tratamiento_3 = round(np.mean(tratamiento_3), 2)
```

```python
print("La media del tratamiento 1 es: ", media_tratamiento_1)
print("La media del tratamiento 2 es: ", media_tratamiento_2)
print("La media del tratamiento 3 es: ", media_tratamiento_3)
```
