```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2

x= np.linspace(-10,10,num=1000)

def df(f): ##Se encarga de realizar la derivada de x
  h=0.000001
  return (f(x+h)-f(x))/h

y= f(x)

plt.plot(x,y,'b')
plt.plot(x,df(f),'r') #Usamos la derivada de de f(x) como Y
plt.grid()
```
