```
import numpy as np
from numpy.random import binomial
from scipy.stats import binom
from math import factorial
import matplotlib.pyplot as plt
import seaborn as sns
```

## Secuencias de eventos tipo Bernoulli

$$
P(k, n; p) = {n \choose k} p^k (1-p)^{n-k} = \frac{n!}{k!(n-k)!}p^k (1-p)^{n-k}
$$

```
def my_binomial(k,n,p):
  return factorial(n)/factorial(k)*(factorial(n-k))*pow(p,k)*pow(1-p,n-k)
```

```
my_binomial(3,5,0.5)
```

```
dist = binom(5,0.5)
dist.pmf(2)
```

## Simulaciones de secuencias con generadores aleatorios

* Los generadores aleatorios tienen como propósito simular muestras de datos que resultarían de muestreos en la vida real de procesos aleatorios como lanzar una moneda o un dado.

```
#Simulación con 100 lanzamientos de moneda equilibrada
p=0.5
n=3
binomial(n,p)
```

```
arr = []
for x in range(100):
  arr.append(binomial(n,p))
  
```

```
def plot_hist(num_trial):
  values = [0,1,2,3] #veces que puede caer cara osea K
  arr = []
  for x in range(num_trial):
    arr.append(binomial(n,p))
  sim = np.unique(arr,return_counts=True)[1]/len(arr) #regresa en porcentaje
  # cuantas veces aparecio el numero de caras en 3 tiros
  teorica = [binom(3,0.5).pmf(k) for k in values] #regresa una lista con los
  #valores teoricos de la distribución
  plt.bar(values,sim,color="green") #dibuja gráfico de barras del caso real
  plt.bar(values,teorica,alpha=0.5,color="blue") #dibuja la teoría
  plt.title("{} Experimentos".format(num_trial)) #cambia el titulo
  plt.show()
```

```
plot_hist(50),plot_hist(150),plot_hist(50000)
```
