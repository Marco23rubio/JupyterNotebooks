```python
import pandas as pd
```

```python
df_books= pd.read_csv("/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv",sep=",",header=0)
df_books.head(5)
```

```python
def two_times(value):
  return value*2
```

```python
df_books["Rating_2"] = df_books["User Rating"].apply(two_times)
df_books.head(5)
```

```python
df_books["Rating_3"] = df_books["User Rating"].apply(lambda x:x*3)
df_books.head(5)
```

```python
df_books["Rating_4"] = df_books.apply(lambda x:x["User Rating"]*2 if x["Genre"]=="Fiction" else x["User Rating"],axis=1)
df_books
```
