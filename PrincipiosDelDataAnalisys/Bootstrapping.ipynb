```python
import pandas as pd
import numpy as np
import random
```

```python
data = np.random.normal(loc=34,size=10000)
##10,000 datos con una media de 34
data
```

```python
data.mean()
```

### Inicio de bootstrapping para calcular el promedio de edades

```python
promedio = []
for _ in range(10000):
  muestra = random.sample(data.tolist(),1000)
  prom = np.mean(muestra)
  promedio.append(prom)
```

```python
np.mean(promedio)
```
