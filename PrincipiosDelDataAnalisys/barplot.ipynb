```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ["India","Japan","Mexico","Colombia","Germany"]
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population,width=0.5,color=["red","blue"],)
plt.xticks(rotation=45)
plt.title("Top 5 countries by Population",color="brown",fontweight="bold")
plt.xlabel("Countries")
plt.ylabel("Population x1000")
plt.legend()
plt.show()
```
