```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.displot(data=tips,x="tip",bins=10,hue="sex",
            stat="count",multiple="dodge")
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",shade=True,
            bw_adjust=2)
plt.show
```
