```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, color="0")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.violinplot(
    data=tips, x="day", y="total_bill", hue="sex", palette="pastel", split=True
)
plt.show()
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex", dodge=True, kind="swarm", col="time"
)
plt.show()
```
