```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip",hue="sex",kind="hist",
              marginal_ticks=True,marginal_kws=dict(bins=10,fill=True,multiple="dodge"))
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="inferno",corner=True,kind="scatter")
plt.show()
```
