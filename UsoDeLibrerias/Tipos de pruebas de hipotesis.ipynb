```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

###Tipos de pruebas

### a)Distribucion de t de student

$$
SE = \frac{\sigma}{\sqrt{N}}
$$

```python

se_length=iris['sepal_length'].std()/np.sqrt(iris.shape[0])
##se usa iris.shape para obtener el numero de filas
se_width=iris['sepal_width'].std()/np.sqrt(iris.shape[0])
```

$$
SED = \sqrt{\frac{\sigma_1^2}{N_1}+\frac{\sigma_2^2}{N_2}}
$$

```python
sed = np.sqrt((se_length**2)+(se_width**2))
```

$$
t = \frac{\bar{x_1}-\bar{x_2}}{SED}=\frac{x_1-x_2}{\sqrt{\frac{(s_1)^2}{n_1}+\frac{(s_2)^2}{n_2}}}
$$

```python
t_value = (iris['sepal_length'].mean() - iris['sepal_width'].mean())/sed
t_value
```

```python
p = ttest_ind(iris['sepal_length'],iris['sepal_width'])
p
```

### b)Correlación y coeficiente de Pearson

```python
iris['sepal_length'].corr(iris['sepal_width'])
```

```python
st.pearsonr(iris['sepal_length'],(iris['sepal_width']))
```

###C) Análisis de Varianza(ANOVA)

```python
f_oneway(iris['sepal_length'], iris['sepal_width'])
```
