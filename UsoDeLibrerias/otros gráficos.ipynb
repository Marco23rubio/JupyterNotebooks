```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data,bins=30,width=1)
plt.show()
```

```python
plt.hist(data,histtype="step")
plt.show()
```

```python
data = np.append(data,200) #Se agrega un 200 dato outlier a comparacion
#del rango original entre 1 a 50
plt.boxplot(data,vert=False,notch=True,showfliers=True)
plt.show()
```

## Scatter

```python
N = 50
x= np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors,marker="^",alpha=0.5)
plt.show()
```
