```python
import pandas as pd
import numpy as np
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded["cancerdata.csv"]))
df.head()
```

```python
x = df.iloc[:,2:] ##Se guardan las variables numericas en una tabla
y = df.iloc[:,1] ##Se guardan la variable categorica
```

```python
x.head()
```

```python
y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

#Se dividel los datos para conjuntos de entrenamiento y prueba con 70% de datos de entrenamiento
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=4)
#Se entrena el modelo
modelo= DecisionTreeClassifier()
modelo.fit(x_train,y_train)
#Se evalua el rendimiento del modelo entre mayor sea el valor mejor
resultado = modelo.score(x_test,y_test)
print(resultado)
```

```python
#Validacion cruzada k fold
from sklearn.model_selection import KFold
##Divide los datos en 10 pliegues como en la imagen
modelo=DecisionTreeClassifier()
kfold_validacion=KFold(10)
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo,x,y,cv=kfold_validacion)
print(resultados)
resultados.mean()
```
