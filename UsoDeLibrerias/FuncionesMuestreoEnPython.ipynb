###Funciones de muestreo

####Muestreo aleatorio simple

```python
aleatorio_8 = econdata.sample(n=6)
aleatorio_8
```

####Muestreo sistematico

```python
def muestreo_sistematica(data,pasos):
  indices = np.arange(0,len(data),step=pasos) ##Aqui ponemos que inicie de fila 0 al largo del DF
##La variable step   indica los brincos que ira dando el nuevo DF
  muestra_sistemaca = data.iloc[indices] ##Guarda el nuevo DF con los indices
  return muestra_sistemaca

muestra = muestreo_sistematica(econdata,3) ##Da brincos de 3 en 3
muestra
```

####Muestreo estratificado

```python
econdata['estratificado'] = econdata['delegacion']+","+econdata['tipo']
round((econdata['estratificado'].value_counts()/len(econdata)).sort_values(ascending=False),2)
```
