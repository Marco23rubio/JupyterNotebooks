```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
```

###Generar una distribuci√≥n

```python
np.random.seed(20)
var1 = np.random.randint(0,10,10)##Dias de vacaciones
var2 = var1 + np.random.normal(0,1,10)#Dinero gastado
```

```python
var1
var2
```

### Visualizacion de datos con un intervalo de confianza

```python
grafico = sns.regplot(x = var1, y = var2, ci=80)
grafico
```

```python
tuple = list(zip(var1,var2))
tuple
```

```python
tabla = pd.DataFrame(tuple,columns=['Dias_vacaciones','Dinero_gastado'])
tabla
```

###Calculo de intervalos de confianza

```python
##95%
valores=st.t.interval(confidence=0.95,df=len(tabla)-1,loc=np.mean(tabla),scale=st.sem(tabla))
valores
```
